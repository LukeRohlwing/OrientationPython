```python
import numpy
import matplotlib.pyplot as plt
```

```python
numpy.random.uniform()
```

```python
uniform = numpy.random.uniform(size = 1000)
```

```python
plt.hist(uniform)
```

```python
poisson = numpy.random.poisson(size = 1000)
```

```python
plt.hist(poisson)
```

```python
binomial = numpy.random.binomial(1, 0.5, size = 1000)
```

```python
plt.hist(binomial)
```

```python
binomial = numpy.random.binomial(100, 0.5, size = 10)
plt.hist(binomial)
```

```python
binomial = numpy.random.binomial(100, 0.5, size = 30)
plt.hist(binomial)
```

```python
binomial = numpy.random.binomial(100, 0.5, size = 10000)
plt.hist(binomial, bins = 14)
```
